```python
import pandas as pd
import numpy as np
from pylatex import Math

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression,LinearRegression, Lasso
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score, GridSearchCV, KFold

import statsmodels.api as sm

import matplotlib.pyplot as plt
import seaborn as sns

from itertools import combinations
```

## 1. Filter the variables {-}

```python
data = pd.read_csv("Covid002.csv", engine='python', encoding='latin1')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
variables_df = pd.read_excel('Variable Description.xlsx')

# Filter Opportunity Insights
grouped = variables_df.groupby('Source')['Variable']
group1 = grouped.get_group('Opportunity Insights')
group1 = group1.values.tolist()

# Filter PM COVID
group2 = grouped.get_group('PM_COVID')
group2 = group2.values.tolist()

# Subset the dataset
general = ['county', 'state', 'deathspc'] 
var = general + group1 + group2
covid = data[var]
covid.head(5)
```

```python
print(len(var1))
print(len(var2))
covid.shape
```

## 2. Compute summary statistics {-}

```python
# Opportunity Insights
covid.describe()
```

## 3. Drop missing values {-}

```python
# Drop all nan observations
covid = covid.dropna(axis=0, how='any')
```

## 4. Create dummy variables {-}

```python
dummies = pd.get_dummies(covid['state'])
dummies.shape
```

```python
# Join the encoded dummies
covid = covid.drop('state',axis = 1)
covid = covid.join(dummies)
covid.head()
```

## 5. Split the sample {-}

```python
X = covid.drop(['county', 'deathspc'], axis = 1)
y = covid['deathspc']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=25)
```

## 6. OLS Estimator {-}

```python
ols_model = LinearRegression().fit(X_train, y_train)
```

### (a) Calculate and report the MSE {-}

```python
# Get the MSE in training sets
y_pred_train = ols_model.predict(X_train)
mse_train = mean_squared_error(y_train, y_pred_train)
print("Training MSE for OLS model:", mse_train)

# Get the MSE in validation sets
y_pred_valid = ols_model.predict(X_test)
mse_valid = mean_squared_error(y_test, y_pred_valid)
print("Validation MSE for OLS model", mse_valid)
```

### (b) Evidence of overfitting {-}

```python
# Get the coefficient of determination - R2
print(ols_model.score(X_train, y_train))
print(ols_model.score(X_test, y_test))
```

#### Answer (b): {-}

Overfitting happens when the statistical model begins to describe the random error in the data rather than the relationships between variables. If the model performs better on the training set than on the test set, it is likely that the model is overfitting.

In this OLS model, there is evidence of overfitting. The inflated R-squared value in training sets is a symptom. We can see that, the R squared score of training sets is approximately 41.08%, while the score of validation sets is only about 36.08%. Besides, the training MSE is much smaller than validation MSE (1561.86 < 1927.25). That is because the model focuses too much on reducing training MSE and works too hard to find patterns in the training data that are just caused by random chance (lower bias but higher variance), leading to the problem of overfitting.

## 7. RR & Lasso {-}

### (a) Define regularization parameters $\lambda$ {-}

```python
# Define range of values of lambda
## a: [-2,2]
## lambda = b = 10^a: [0.01,100]
a = np.linspace(start=-2, stop=2, num=100)
b = 10**a
alpha_param = b

# Create a parameters grid
param_grid = [{'alpha': alpha_param }]
```

### (b) Use 10FCV to estimate test errors {-}

#### Ridge {-}

```python
from sklearn.linear_model import Ridge
ridge = Ridge(normalize = True)

# cv = 10
grid_search_ridge = GridSearchCV(ridge, param_grid, cv = 10, scoring = 'neg_mean_squared_error')
grid_search_ridge.fit(X_train, y_train)

# Calculate a vector of mean and standard deviation values for each lambda
def vector_values(grid_search, trials):
    mean_vec = np.zeros(trials)
    std_vec = np.zeros(trials)
    i = 0
    final = grid_search.cv_results_
    
    for mean_score, std_score in zip(final["mean_test_score"], final["std_test_score"]):
        mean_vec[i] = - mean_score
        std_vec[i] = std_score
        i = i+1
        
    return mean_vec, std_vec

# Test error for each Ridge model
mean_vec, std_vec = vector_values(grid_search_ridge, 100)
```

#### Lasso {-}

```python
from sklearn.linear_model import Lasso
lasso = Lasso(normalize = True)

# cv = 10
grid_search_lasso = GridSearchCV(lasso, param_grid, cv = 10, scoring = 'neg_mean_squared_error')
grid_search_lasso.fit(X_train, y_train)

# Test error for each Lasso model
mean_vec_l, std_vec_l = vector_values(grid_search_lasso, 100)
```

### (c) Plot the 10FCV estimates of test errors {-}

#### Ridge {-}

```python
# MSE vs. log(lambda) = a: [-2,2]
plt.figure(figsize=(12,10))
plt.title('Ridge Regression', fontsize=16)
plt.plot(np.log(alpha_param), mean_vec)
plt.errorbar(np.log(alpha_param), mean_vec, yerr = std_vec)
plt.ylabel("MSE", fontsize=16)
plt.xlabel("log(lambda)", fontsize=16)
plt.show()
```

#### Lasso {-}

```python
plt.figure(figsize=(12,10))
plt.title('Lasso Regression', fontsize=16)
plt.plot(np.log(alpha_param), mean_vec_l)
plt.errorbar(np.log(alpha_param), mean_vec_l, yerr = std_vec)
plt.ylabel("MSE", fontsize=16)
plt.xlabel("log(lambda)", fontsize=16)
plt.show()
```

### (d) Choose the optimal value of $\lambda$ {-}

#### Ridge {-}

```python
# Find the lowest MSE score
print("lowest MSE score for Ridge model:", min(mean_vec))

# Optimal lambda: one that minimizes MSE
lambda_r = alpha_param[np.where(mean_vec == min(mean_vec))][0]
print("the optimal value of lambda for Ridge model:", lambda_r)
```

#### Lasso {-}

```python
# Find the lowest MSE score
print("lowest MSE score for Lasso model:", min(mean_vec_l))

# Optimal lambda: one that minimizes MSE
lambda_l = alpha_param[np.where(mean_vec_l == min(mean_vec_l))][0]
print("the optimal value of lambda for Lasso model:", lambda_l)
```

### (e) Re-estimate the model {-}

#### Ridge {-}

```python
# Re-esitimate the Ridge model using lambda_r
ridge_model = Ridge(alpha=lambda_r, fit_intercept=True, 
                    normalize=True, max_iter=1000000).fit(X_train, y_train)

df1 = pd.DataFrame({'feature': X.columns, 'coefficient': ridge_model.coef_})
df1.insert(loc=0, column='index1', value=np.arange(len(df1)))
display(df1)
```

```python
# Create a plot
plt.figure(figsize=(16,12))
ax = sns.barplot(x='index1', y='coefficient', color="#4169E1", data=df1)
ax.set_title('Coefficient of Ridge model with lambda_r', fontsize=16)
plt.show()
```

#### Lasso {-}

```python
# Re-esitimate the Lasso model using lambda_l
lasso_model = Lasso(alpha=lambda_l, fit_intercept=True, 
                    normalize=True, max_iter=1000000).fit(X_train, y_train)

df2 = pd.DataFrame({'feature': X.columns, 'coefficient': lasso_model.coef_})
## Drop the predictors that coefficient = 0
df2 = df2[df2['coefficient'] != 0]
df2.insert(loc=0, column='index1', value=np.arange(len(df2)))
display(df2)
```

```python
# Create a plot
plt.figure(figsize=(16,12))
ax = sns.barplot(x='index1', y='coefficient', color="#008B8B", data=df2)
ax.set_title('Coefficient of Lasso model with lambda_l', fontsize=16)
plt.show()
```

### Answer (e): {-}

Above are the re-esitmation results of Ridge and Lasso models using optimal value of $\lambda$. The Ridge model doesn't do model selection, but includes all p regressors because the penalty term doesnt send any coefficients to exactly zero. However, compared to all 107 predictors selected in the Ridge model, the Lasso model does model selection with only 64 predictors selected.

## 8. MSE for Ridge/Lasso/OLS {-}

#### Ridge {-}

```python
ridge_model = Ridge(alpha=lambda_r, fit_intercept=True, 
                    normalize=True, max_iter=1000000).fit(X_train, y_train)

# training set
ridge_tpre = ridge_model.predict(X_train)
tmse_r = mean_squared_error(y_train, ridge_tpre)
print("Training MSE for Ridge model:", tmse_r)

# validation set
ridge_pre = ridge_model.predict(X_test)
vmse_r = mean_squared_error(y_test, ridge_pre)
print("Validation MSE for Ridge model:", vmse_r)
```

#### Lasso {-}

```python
lasso_model = Lasso(alpha=lambda_l, fit_intercept=True, 
                    normalize=True, max_iter=1000000).fit(X_train, y_train)

# training set
lasso_tpre = lasso_model.predict(X_train)
tmse_l = mean_squared_error(y_train, lasso_tpre)
print("Training MSE for Lasso model:", tmse_l)

# validation set
lasso_pre = lasso_model.predict(X_test)
vmse_l = mean_squared_error(y_test, lasso_pre)
print("Validation MSE for Ridge model:", vmse_l)
```

#### OLS {-}

```python
# training set
ols_tpre = ols_model.predict(X_train)
tmse_o = mean_squared_error(y_train, ols_tpre)
print("Training MSE for OLS model:", tmse_o)

# validation set
ols_pre = ols_model.predict(X_test)
vmse_o = mean_squared_error(y_test, ols_pre)
print("Validation MSE for OLS model", vmse_o)
```

### Answer 8: {-}

Whether choosing OLS or Ridge/Lasso is a trade-off between bias and variance. In general, the penalties of Ridge regression and Lasso analogs reduce flexibility, increasing bias and lowering variance. From the results, we can see that the validation MSE for both Ridge and Lasso are lower than OLS (VMSE: Lasso < Ridge < OLS) by sacrificing the training MSE (TMSE: OLS < Ridge < Lasso). Hence, both Ridge and Lasso improve the prediction over OLS by reducing overfitting (lower variance), and Lasso model performs the best with the lowest test MSE.

Thus, I would recommend Lasso model to the CDC. On the one hand, it performs best on out-of-sample predictions. On the other hand, since only 64 out of 107 predictors are selected in the model, it will also be easier for researchers to interpret the relationships between predictors and responce, as well as prevent overfitting.
